## Import OpenCV Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def print_cv(image , title="Image"):
    # Display the image using Matplotlib
    plt.title(title)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')  # Turn off axis labels
    plt.show()
```

```python
def show_cv(image , title="Image" , end_time=0 ):
    # Display the image using matplotlib
    cv2.imshow(title, image)
    cv2.waitKey(end_time)
    cv2.destroyAllWindows()
```

```python
def read_cv(image_location , image_mode = -1):
    return cv2.imread(image_location , image_mode)
```

```python
def write_cv(image_location:str , image):
    return cv2.imwrite(image_location , image)
```

```python
def resize_cv(image , width , height):
    return cv2.resize(image , (width , height))
```

```python
def rotate_cv(image , rotation:int = 0):
    rotation = rotation % 4 - 1
    if rotation == -1:
        return image
    return cv2.rotate(image , rotation)
```

## Capture Video

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    cv2.imshow("Video", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
def cap_video_cv(sourse , live = True):
    cap = cv2.VideoCapture(sourse)
    while True:
        ret, frame = cap.read()
        
        cv2.imshow("Video", frame)
        if cv2.waitKey(1) == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
# Plot for F at distance x
a = 0.9
e = 8.85*10**-12
q = 2*10**-5
# create a list of x values frm 0 to 10 with step 0.1
x = np.arange( 0 , 10 , 0.01)
f = q*q*x / ( 4 * np.pi * e * (( x**2 + a**2 )**(3/2)) )
plt.plot(x , f)
plt.axhline(y=1, color='red', linestyle='--', label='f = 1')
plt.axvline(x=0.25, color='green', linestyle='--', label='x')
plt.axvline(x=1.5, color='green', linestyle='--', label='x')

plt.axvline(x=.9, color='purple', linestyle='--', label='x')
```

```python
f = lambda v: 0.01*v**7 + 0.05*v**5 + 0.1*v**3 + 2*v - 2
f_dif = lambda v: 7*0.01*v**6 + 5*0.05*v**4 + 3*0.1*v**2 + 2

v = 2
print("v0 = " , v)
for i in range(20):
    v = v - f(v)/f_dif(v)
    print("v" , i+1 , " = " , v)
```
